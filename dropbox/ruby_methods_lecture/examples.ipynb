### define a method that takes any number of arguments and returns their sum

```ruby
def random_number
    (0..1000).to_a.sample
end
```

```ruby
random_number
```

```ruby
def sum(*numbers)
    numbers.sum
end
```

```ruby
def lots_of_numbers
    random_number.times.map do 
        random_number
    end
end
```

```ruby
def test(expected_answer, *numbers)
    sum(*numbers) == expected_answer
end
```

```ruby
test(7, 3, 4)
```

```ruby
test_data = 10.times.map do
    xs = lots_of_numbers
    answer = xs.sum
    [answer, xs]
end
```
